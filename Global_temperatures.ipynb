```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
data = pd.read_excel("GlobalTemperatures.csv.xlsx")
data = data.dropna()
data['dt'] = pd.to_datetime(data['dt'])
data['Year'] = data['dt'].dt.year

# Yearly average temperature
yearly_avg = data.groupby('Year')['LandAverageTemperature'].mean()

# Line Chart
plt.figure(figsize=(12,10))
plt.plot(yearly_avg.index, yearly_avg.values, color='red')
plt.title("Global Average Temperature Change Over Decades")
plt.xlabel("Year")
plt.ylabel("Temperature (°C)")
plt.show()   # gap

# Bar Chart (Decade-wise)
decade = (data['Year']//10)*10
decade_avg = data.groupby(decade)['LandAverageTemperature'].mean()

plt.figure(figsize=(12,6))
plt.bar(decade_avg.index, decade_avg.values, color='orange')
plt.title("Average Temperature by Decade")
plt.xlabel("Decade")
plt.ylabel("Temperature (°C)")
plt.show()   # gap

# Histogram
plt.figure(figsize=(10,6))
plt.hist(data['LandAverageTemperature'], bins=30, color='blue', edgecolor='black')
plt.title("Distribution of Global Temperatures")
plt.xlabel("Temperature (°C)")
plt.ylabel("Frequency")
plt.show()   # gap

# Boxplot
plt.figure(figsize=(12,6))
sns.boxplot(x=decade, y=data['LandAverageTemperature'])
plt.title("Temperature Variation by Decade")
plt.xlabel("Decade")
plt.ylabel("Temperature (°C)")
plt.show()   # gap

# Scatter Plot
plt.figure(figsize=(12,6))
plt.scatter(data['Year'], data['LandAverageTemperature'], alpha=0.5, color='green')
plt.title("Scatter Plot of Temperatures Over Years")
plt.xlabel("Year")
plt.ylabel("Temperature (°C)")
plt.show()   # gap

# Heatmap
plt.figure(figsize=(6,8))
sns.heatmap(decade_avg.values.reshape(-1,1), annot=True, cmap="coolwarm")
plt.title("Average Temperature by Decade (Heatmap)")
plt.show()   # gap
```
